# DocTutor

Tiny Transformer encoder retrieval chatbot over the eight TensorTonic Transformer lessons. The browser app uses `doctutor.onnx`, `vocab.json`, `chunks.json`, and `chunk_embeddings.json`.


## Pipeline

```text
User question
  -> tokenizer
  -> embedding layer
  -> positional encoding
  -> transformer encoder block(s)
  -> pooled query vector
  -> compare against lesson chunk vectors
  -> retrieve best chunks
  -> show short answer + citation
```

## Model

- Word-level tokenizer with fixed `<PAD>`, `<UNK>`, `<BOS>`, `<EOS>` IDs.
- Scaled token embeddings plus sinusoidal positional encodings.
- Two small post-norm Transformer encoder blocks.
- Mean pooling over non-padding tokens.
- Contrastive retrieval training against lesson chunks.
